```python
%autosave 120
import pandas as pd 
import numpy as np 
```

## Importing the data

```python
df_raw = pd.read_csv('NBA_Season_Data.csv')
```

#### Cleaning the data from N/A and zero values

```python
df = df_raw.dropna(subset=['Height', 'TrueSalary']) 
df = df[df.Height != 0]
```

#### Cleaning the Salary data from the $ sign

```python
df.TrueSalary =df.TrueSalary.apply(lambda x: x.replace('$',''))
df.TrueSalary =df.TrueSalary.apply(lambda x: x.replace(',',''))
df.TrueSalary =df.TrueSalary.apply(lambda x: float(x))
```

## Describtive statistics

### Height data analysis

```python
df.hist(column='Height', bins=20)
print(df['Height'].describe())
print('mean = ',df['Height'].mean())
print('median = ',df['Height'].median())
print('mode = ',df['Height'].median())
```

### Salary data analysis

```python
df.hist(column='TrueSalary', bins=20)
print(df['TrueSalary'].describe())
print('mean = ',df['TrueSalary'].mean())
print('median = ',df['TrueSalary'].median())
print('mode = ',df['TrueSalary'].median())
```
